# **OPTIMIZATION PROBLEM üéØ**


```python
### üéØ ‚ûï üîΩOptimization Problems
### Optimization is a branch of mathematics that deals with finding the best solution
### to a problem by minimizing or maximizing a function.
### A general optimization problem is defined as:
###
###     min f(x) subject to x ‚àà C ‚äÜ R^k
###
### where:
### - `f(x)` is the objective function to be minimized.
### - `C` is the constraint set where the solution must lie.
### - `x` is the decision variable.
###
### If `C` equals the entire space `R^k`, the problem is **unconstrained**.
### If there are constraints such as `g_i(x) ‚â§ 0` or `h_j(x) = 0`, the problem is **constrained**.
###
### Constrained optimization problems arise frequently in economics, engineering,
### and machine learning, where feasible solutions must satisfy physical, financial,
### or logical constraints.
###
### For now, we focus on unconstrained optimization where `f(x)` is smooth
### (i.e., continuously differentiable everywhere).


### üîìüÜìüåê Unconstrained Problems
### In an unconstrained problem, we aim to find a global minimizer `x*`, which satisfies:
###
###     f(x*) ‚â§ f(x) for all x in R^k
###
### However, finding a **global minimizer** can be difficult because we only have
### local information about `f(x)`. Thus, most algorithms only find **local minimizers**,
### which satisfy:
###
### - `x*` is a **local minimizer** if there exists a neighborhood `N` around `x*` such that:
###     f(x*) ‚â§ f(x) for all x in N.
### - `x*` is a **strict local minimizer** if:
###     f(x*) < f(x) for all x in N \ {x*}.
###
### üîçüîçüîç Example:
### - The function f(x) = x¬≥ has a local minimum at x = 0.
### - The function f(x) = (x - a)¬≤ has a strict local minimum at x = a.
###
### Global minimizers are difficult to guarantee in non-convex problems. In practice,
### methods such as genetic algorithms are sometimes used to approximate global solutions.

### ‚ûó ‚úñÔ∏è ‚ûï üßÆ Fundamental Mathematical Tools
### To analyze optimization problems, we use:
###
### 1Ô∏è‚É£. **Gradient (First-Order Derivative)**
### The gradient `‚àáf(x)` of a function f(x) at a point x‚ÇÄ is:
###
###     ‚àáf(x‚ÇÄ) = ( ‚àÇf/‚àÇx‚ÇÅ, ‚àÇf/‚àÇx‚ÇÇ, ..., ‚àÇf/‚àÇx_k )
###
### It represents the direction of the steepest increase of f(x).
### Gradient descent algorithms use `-‚àáf(x)` to iteratively find minima.
###
### 2Ô∏è‚É£. **Hessian Matrix (Second-Order Derivatives)**
### The Hessian `‚àá¬≤f(x)` is a k√ók matrix of second-order partial derivatives:
###
###     ‚àá¬≤f(x‚ÇÄ) = [ ‚àÇ¬≤f/‚àÇx·µ¢‚àÇx‚±º ] evaluated at x‚ÇÄ
###
### The Hessian helps determine the curvature of f(x):
### - If `‚àá¬≤f(x*)` is **positive definite**, x* is a strict local minimizer.
### - If `‚àá¬≤f(x*)` is **negative definite**, x* is a strict local maximizer.
### - If `‚àá¬≤f(x*)` has both positive and negative eigenvalues, x* is a saddle point.
###
### 3Ô∏è‚É£. **Necessary and Sufficient Conditions for Optimality**
### - **First-Order Condition**: A local minimizer x* satisfies `‚àáf(x*) = 0`.
### - **Second-Order Condition**: If `‚àá¬≤f(x*)` is positive definite, then x* is a local minimum.
### - If `‚àá¬≤f(x*)` is semi-positive definite, x* may still be a local minimum,
###   but additional conditions must be checked.
###
### What does **semi-positive definite** mean?
### - A Hessian matrix `H` is **semi-positive definite** if for all vectors `z`, we have:
###
###     z·µÄ H z ‚â• 0
###
### - This means that `H` does not have any strictly negative eigenvalues, but it may
###   have some zero eigenvalues.
### - When `‚àá¬≤f(x*)` is semi-positive definite, `x*` could be a local minimum, but not necessarily strict.
### - If the Hessian is positive definite (`z·µÄ H z > 0` for all z ‚â† 0)
###
### 4Ô∏è‚É£. **Convex Functions and Global Optimality**
### If f(x) is **convex**, then any local minimizer is also a global minimizer.
### For differentiable convex functions, any point where `‚àáf(x) = 0` is a global minimizer.
###
### üìùüìùüìùExample:
### - The function `f(x) = x¬≤` is convex, and its unique minimizer is at x = 0.
### - The function `f(x) = e^x` is convex but has no minimizer over R.
### - The function `f(x) = x¬≥` is not convex and has local extrema.



###Now we see how the computer solves those type of problems
### ü§ñ üîÑ üß† Overview of Algorithms
### Optimization algorithms aim to find a local minimum (or maximum) of a function by iteratively updating a candidate solution.
### These algorithms typically improve a starting point `x(0)` by moving in a direction that decreases the function value.
###
###‚öôÔ∏è‚öôÔ∏è‚öôÔ∏è
### One common approach is **descent direction methods**, which generate a sequence of points that converge to a local minimum.
### A widely used descent method is **Gradient Descent**, which updates `x` using the formula:
###
###     x(t+1) = x(t) - Œ± ‚àáf(x(t))
###
### where:
### - `‚àáf(x(t))` is the gradient of the function at `x(t)`, indicating the direction of steepest ascent.
### - `Œ±` (step size or learning rate) controls the step magnitude in the descent direction.
###
### The intuition behind gradient descent is that moving in the opposite direction of the gradient leads toward a minimum.
### If `Œ±` is too large, the algorithm may overshoot the minimum; if `Œ±` is too small, convergence will be slow.
###
### The stopping criteria for these algorithms are usually:
### 1Ô∏è‚É£. **Maximum iterations reached**: The algorithm stops after a predefined number of steps.
### 2Ô∏è‚É£. **Absolute improvement**: If the change in function value is smaller than a threshold.
### 3Ô∏è‚É£. **Relative improvement**: The function value changes less than a given fraction of the previous step.
### 4Ô∏è‚É£. **Gradient magnitude**: The algorithm stops when `||‚àáf(x)|| < Œµ`, meaning the gradient is close to zero.
###
### Detailed Explanation of Gradient Descent üß†
### - The computer needs to be **instructed on the function** `f(x)` to be minimized and its derivative `‚àáf(x)`.
### - Once these are defined, the algorithm starts from an **initial guess** `x(0)`.
### - At each step, the computer calculates the gradient and updates the current value of `x` accordingly.
### - The process continues until one of the stopping conditions is met.
###
### üö´üö´üö´ Stopping Conditions in More Detail:
### - **Maximum Iterations**: The algorithm runs for a fixed number of steps and stops regardless of whether it has converged.
### - **Function Improvement Threshold**: If `|f(x_t) - f(x_{t+1})|` is smaller than a predefined tolerance, the algorithm assumes it has reached the minimum.
### - **Relative Function Improvement**: The function value difference is considered relative to its previous value.
### - **Gradient Magnitude**: When `||‚àáf(x)||` is very small, it suggests that `x` is close to an optimal solution.
###
```

```python
# Define the function f(x) = x^2 - log(x) and its derivative
def f(x):
    """
    Function to be minimized: f(x) = x^2 - log(x)
    The computer needs to know this function explicitly.
    """
    return x**2 - np.log(x)

def df(x):
    """
    Derivative of the function: df(x)/dx = 2x - 1/x
    This tells the algorithm the direction in which to move.
    """
    return 2*x - 1/x

def connectpoints(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'k-')

# Initialize parameters
x0 = 2   # Initial guess
alpha = 0.3  # Step size (learning rate)

# Perform gradient descent updates
x1 = x0 - alpha * df(x0)  # First update
x2 = x1 - alpha * df(x1)  # Second update

# Generate points for plotting the function
# Points for the function plot
x = np.linspace(-2.5, 2.5, 400)
y = f(x)

# Tangent line at x0 (y = m*x + b)

# Creating the plot

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x^2')
plt.scatter([x0, x1], [f(x0), f(x1)], color='red')  # Points
m = df(x0)
b = f(x0) - m*x0
tangent_line = m*x + b
plt.plot(x, tangent_line, 'b--', label=f'Tangent at x0={x0}')
plt.arrow(x0, 0.4, x1-x0, 0.0, head_width=0.1, length_includes_head=True, color = 'r')

m = df(x1)
b = f(x1) - m*x1
tangent_line = m*x + b
plt.plot(x, tangent_line, 'b--', label=f'Tangent at x0={x1}', )
plt.ylim([-0.2,6])
plt.xlim([-0.4,3])
plt.plot(x, tangent_line, 'b--', label=f'Tangent at x0={x1}', )
m = df(x0)
b = f(x0) - m*x0
tangent_line = m*x + b

plt.scatter(x0, f(x0), color='green')  # Initial point
plt.scatter(x1, f(x1), color='green')  # Next point after step
plt.scatter(x2, f(x2), color='green')  # Next point after step

plt.arrow(x1, 0., x2-x1, 0., head_width=0.1, length_includes_head=True, color = 'r')

plt.title('Gradient Descent on f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xticks([])  # Remove x-axis ticks
plt.xticks([x0, x1, x2], [r"$x^{(0)}$", r"$x^{(1)}$", r"$x^{(2)}$"])
#plt.legend()
plt.grid(True)
plt.plot([x0, x0],[f(x0), 0],'g--')
plt.plot([x1, x1],[f(x1), 0],'g--')
plt.plot([x2, x2],[f(x1), 0],'g--')
plt.show()

### üìùüìàüìâExplanation of the Graph
### The plot visualizes the function `f(x) = x^2 - log(x)` and the gradient descent steps.
### - The red points indicate the sequence of x-values obtained via gradient descent.
### - The dashed blue lines are tangents at `x0` and `x1`, showing the slope (gradient direction).
### - The arrows represent the movement in the negative gradient direction.
###
### üëÄüëÄüëÄObservations:
### - Initially, at `x0 = 2`, the gradient is positive, meaning `f(x)` is increasing. The algorithm moves left.
### - The step size `Œ± = 0.3` determines how far the next `x` is from the current one (learning rate).
### - After two iterations, the descent brings `x` closer to the minimum.
### - The gradient decreases in magnitude, meaning the function value is stabilizing, indicating convergence.
###
### This graph confirms the behavior of the gradient descent algorithm:
### it moves toward a local minimum while reducing function values at each step.
###
### üîëüîëüîëKey Takeaways:
### - The choice of step size `Œ±` is crucial: too large can cause oscillations, too small results in slow convergence.
### - The gradient at each step gives the best direction to move toward the minimum.
### - Stopping criteria ensure that the algorithm terminates efficiently without unnecessary computations.

#ü§îü§îü§î An intereresting thing to visualize is how different criteria affect the final point the algoritm finds
#Now we impelement a code that provides different graphs with the different stopping criteria
#Let's specifiy that in the previous code the stopping criterion is not explicitly implemented. The script performs only two iterations of gradient descent manually:
## x1 = x0 - alpha * df(x0)  # First update
## x2 = x1 - alpha * df(x1)  # Second update

### Stopping Criteria in Gradient Descent (As I wrote before)
### In optimization algorithms like gradient descent, choosing an appropriate stopping criterion
### is crucial to balance computational efficiency and accuracy. There are four main stopping conditions:
###
### 1Ô∏è‚É£ **Maximum Iterations Reached**
###    - The algorithm stops after a fixed number of steps.
###    - Simple but might stop too early or run unnecessarily long.
###
### 2Ô∏è‚É£ **Absolute Improvement**
###    - Stops when the change in function value is below a threshold `Œµ_a`.
###    - Prevents unnecessary iterations when close to a minimum.
###
### 3Ô∏è‚É£ **Relative Improvement**
###    - Stops when the relative change in function value is small (`Œµ_r`).
###    - Useful when function values vary significantly in magnitude.
###
### 4Ô∏è‚É£ **Gradient Magnitude**
###    - Stops when `||‚àáf(x)|| < Œµ_g`, meaning the slope is nearly flat.
###    - The most precise method for finding a local minimum.
```

#**DIFFERENT STOPPING CONDITIONSüö´**

```python
# Define the function and its derivative
def f(x):
    return x**2 - np.log(x)

def df(x):
    return 2*x - 1/x

# Gradient Descent Implementation with Different Stopping Criteria
def gradient_descent(stop_condition, epsilon=1e-6, max_iter=100):
    x = 2  # Initial guess
    alpha = 0.3  # Step size
    history = [x]

    for i in range(max_iter):
        grad = df(x)
        x_new = x - alpha * grad
        history.append(x_new)

        # Stopping Conditions
        if stop_condition == 'max_iter' and i >= max_iter - 1:
            break
        if stop_condition == 'absolute_improvement' and abs(f(x_new) - f(x)) < epsilon:
            break
        if stop_condition == 'relative_improvement' and abs(f(x_new) - f(x)) / abs(f(x)) < epsilon:
            break
        if stop_condition == 'gradient_magnitude' and abs(grad) < epsilon:
            break

        x = x_new

    return np.array(history)

# Run gradient descent for each stopping criterion
methods = ['max_iter', 'absolute_improvement', 'relative_improvement', 'gradient_magnitude']
histories = {method: gradient_descent(method) for method in methods}

# Plot results
plt.figure(figsize=(12, 8))
x_vals = np.linspace(0.5, 2.5, 400)
y_vals = f(x_vals)

for i, method in enumerate(methods):
    plt.subplot(2, 2, i+1)
    plt.plot(x_vals, y_vals, label='f(x)')  # Function curve
    plt.scatter(histories[method], f(histories[method]), color='red', label='Steps')  # Iteration points
    plt.scatter(histories[method][-1], f(histories[method][-1]), color='blue', s=100, label='Final Point')  # Final point in blue
    plt.title(f'Stopping Condition: {method}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

#With this commands I also have underlined the final poin with a different color

### ‚öñÔ∏è‚öñÔ∏è‚öñÔ∏èComparison of Different Stopping Criteria
### - **Maximum Iterations**: The algorithm runs for a fixed number of steps, which may lead to unnecessary iterations.
### - **Absolute Improvement**: Stops when function changes are small, leading to a balance between speed and accuracy.
### - **Relative Improvement**: Stops when function value changes proportionally to previous values, making it scale-aware.
### - **Gradient Magnitude**: The most reliable method, ensuring we stop close to a true local minimum. Stops when slope is nearly zero, typically leading to the best local minimum.
###
### Observations:
### - The stopping points differ depending on the method, affecting the final x-values.
### - "Gradient Magnitude" generally finds the best minimum, while "Max Iterations" may not be efficient.
### - "Absolute" and "Relative Improvement" provide adaptive stopping, balancing performance and accuracy.
###
### This comparison helps understand how different stopping criteria affect optimization outcomes!
```

```python
# Theory: Common Global Optimization Methods
# ---------------------------------------------------------------------------
# In many real-world problems, optimization tasks may involve functions with
# multiple local minima, making it challenging to find the global minimum.
# To address this, various global optimization methods can be used to explore
# the function space more thoroughly and avoid getting stuck in local minima.
#
# Here are some common global optimization techniques:
# ---------------------------------------------------------------------------

# 1. **Multiple Restarts**: üîÑüéØ
#    - Gradient descent is a popular method for finding local minima. However,
#      it is prone to getting stuck in local minima depending on the starting
#      point.
#    - **Multiple Restarts** refers to running the gradient descent algorithm
#      from different initial starting points (random or based on some strategy)
#      to explore different parts of the function's domain.
#    - The best result from all runs is considered as the global minimum.
#    - This method helps explore multiple regions of the function to increase
#      the likelihood of finding the global minimum.

# 2. **Simulated Annealing**: üî•üå°Ô∏è-
#    - It's a probabilistic optimization technique inspired
#      by the physical process of heating and cooling a material to remove defects.
#    - It allows "uphill" moves (i.e., accepting solutions that are worse than
#      the current one) with a certain probability, which helps escape local minima.
#    - Over time, the probability of accepting worse solutions decreases, allowing
#      the algorithm to converge toward the global minimum.
#    - The temperature parameter controls how much randomness is allowed in the
#      process. High temperatures allow more exploration, while lower temperatures
#      reduce randomness and lead to convergence.

# 3. **Genetic Algorithms**: üß¨üå±
#    - The algorithm applies **selection** to choose the best solutions, **crossover**
#      to combine features of good solutions, and **mutation** to introduce diversity
#      and explore new areas of the solution space.
#    - The idea is to explore large areas of the solution space and find better
#      solutions over time through evolutionary processes.

# 4. **Grid Search**: üó∫Ô∏èüî≤
#    - It's a brute-force optimization method where the search space
#      is divided into a grid of values, and the function is evaluated at each
#      point on the grid.
#    - This method is often used when the solution space is small and well-defined,
#      as it guarantees finding the global minimum (if the grid is fine enough).
#    - However, it can be computationally expensive when the search space is large
#      because it involves evaluating the function at many different points.
# ---------------------------------------------------------------------------
# In the following example, we use **Random Restarts** as method for global optimization
# because it is simple and effective. By running the gradient descent from different
# starting points, we increase our chances of finding the global minimum rather than
# being stuck in a local minimum. We will explore this method in the code implementation.

# ---------------------------------------------------------------------------
# üß©üß©üß©Python Code Implementation:
# ---------------------------------------------------------------------------

# Define the function and its derivative
def f(x):
    """
    This is the function we want to minimize. It has multiple local minima and
    is asymmetric.
    """
    return np.sin(2*x) + 0.3*x**2 - 0.1*x**3  # Function with asymmetry and local minima

def df(x):
    """
    The derivative of the function, which is used to calculate the gradient
    for the gradient descent algorithm.
    """
    return 2*np.cos(2*x) + 0.6*x - 0.3*x**2  # Derivative of the function

# Gradient Descent Implementation
def gradient_descent(starting_point, alpha=0.1, epsilon=1e-4, max_iter=100):
    """
    Perform gradient descent from a given starting point to find the local minimum.
    This function stops when the improvement in function value is smaller than epsilon.

    Args:
    - starting_point: Initial guess for the minimum.
    - alpha: Step size (learning rate).
    - epsilon: Threshold for stopping condition.
    - max_iter: Maximum number of iterations.

    Returns:
    - x_min: The point where the gradient descent stops.
    """
    x = starting_point
    for i in range(max_iter):
        grad = df(x)  # Compute gradient (derivative of f)
        x_new = x - alpha * grad  # Update the current guess

        # Stop if the improvement in function value is smaller than epsilon
        if abs(f(x_new) - f(x)) < epsilon:
            break
        x = x_new  # Update x to the new value

    return x

# Global Optimization via Random Restarts
def find_global_minimum(num_restarts=10, alpha=0.1):
    """
    Use multiple random restarts of the gradient descent algorithm to find the global
    minimum. The function starts gradient descent from several random starting points
    to explore different regions of the function.

    Args:
    - num_restarts: The number of random restarts to perform.
    - alpha: Step size (learning rate) for gradient descent.

    Returns:
    - min_x: The x-coordinate of the global minimum.
    - min_value: The corresponding value of the function at the global minimum.
    """
    # Randomly initialize starting points within the range [-3, 3]
    starting_points = np.random.uniform(-3, 3, num_restarts)
    min_value = float('inf')  # Start with a very large minimum value
    min_x = None  # Variable to store the global minimum x-coordinate

    # Run gradient descent from each random starting point
    for start in starting_points:
        x_min = gradient_descent(start, alpha)  # Run gradient descent from this starting point
        if f(x_min) < min_value:  # Update if we find a new global minimum
            min_value = f(x_min)
            min_x = x_min

    return min_x, min_value  # Return the global minimum point and its value

# Identify local minima within a range using scipy's minimize function
def find_local_minima(start_range=-3, end_range=3, step=0.5):
    """
    Find local minima by using scipy's minimize function in a specified range.
    We use a step to explore different regions and find local minima.

    Args:
    - start_range: Start of the range to search for local minima.
    - end_range: End of the range to search for local minima.
    - step: Step size for the grid search.

    Returns:
    - local_minima: List of x-coordinates of local minima.
    - local_minima_values: Corresponding function values at the local minima.
    """
    local_minima = []
    local_minima_values = []

    # Search for local minima in the specified range
    for x0 in np.arange(start_range, end_range, step):
        res = minimize(f, x0)  # Use scipy minimize function to find the local minimum
        if res.success:
            local_minima.append(res.x[0])
            local_minima_values.append(f(res.x[0]))

    return local_minima, local_minima_values

# Find the global minimum using multiple restarts
global_min_x, global_min_value = find_global_minimum(num_restarts=20)

# Find local minima within a range
local_minima_x, local_minima_y = find_local_minima(start_range=-3, end_range=3, step=0.5)

# Plot the function and both global and local minima
x_vals = np.linspace(-3, 3, 400)  # Create an array of x values for plotting
y_vals = f(x_vals)  # Compute the corresponding y values of the function

plt.figure(figsize=(12, 6))

# First plot: showing the global minimum
plt.subplot(1, 2, 1)
plt.plot(x_vals, y_vals, label='f(x)', color='black')  # Function curve
plt.scatter(global_min_x, global_min_value, color='blue', s=100, label=f'Global Minimum ({global_min_x:.4f}, {global_min_value:.4f})')
plt.title('Global Minimum Search with Gradient Descent (Multiple Restarts)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Second plot: showing both global and local minima
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_vals, label='f(x)', color='black')  # Function curve
plt.scatter(global_min_x, global_min_value, color='blue', s=100, label=f'Global Minimum ({global_min_x:.4f}, {global_min_value:.4f})')
plt.scatter(local_minima_x, local_minima_y, color='red', label='Local Minima', zorder=5)
plt.title('Global and Local Minima')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Print the global minimum coordinates and function value
print(f"Global minimum found at x = {global_min_x:.4f} with f(x) = {global_min_value:.4f}")


# ---------------------------------------------------------------------------
# üëÄüëÄüëÄ Key Observations:
# ---------------------------------------------------------------------------
#Theory:
#- In convex functions, the second derivative is always non-negative (f''(x) >= 0), ensuring that the function is always either increasing or decreasing smoothly, forming a "U" shape.
#- For convex functions, if the gradient (first derivative) is zero, we can confirm that the point is a global minimum, and there is no risk of being stuck in a local minimum.
#- In non-convex functions, we cannot rely solely on the first derivative (gradient) to find the global minimum. Gradient descent can lead us to **local minima**, but it might miss the global minimum. In this case, we might use techniques like **multiple restarts** or **simulated annealing** to increase our chances of finding the global minimum.

#Explanation of second derivatives in the non-convex case:
#- For non-convex functions, the second derivative may be negative (f''(x) < 0), indicating the presence of local maxima or points of inflection. Even if we find a point where the gradient is zero, it could be a **local minimum**, a **local maximum**, or a **saddle point**.
#- The second derivative is not always reliable for determining the global minimum, especially when there are **multiple local minima**.

#In our case, gradient descent on the **convex function** is straightforward, as the second derivative is positive and constant, confirming that any local minimum we find is the global minimum.
#For the **non-convex function**, gradient descent might find a local minimum, but we need to perform multiple restarts or use more sophisticated methods (like simulated annealing) to ensure we find the global minimum.

#Additional Consideration for Non-Convex Functions:
#- If we have two **local minima** in a non-convex function, the second derivative could still be positive at both points (indicating they are both local minima), but one of them could be the **global minimum**.
#- We can't always rely on the second derivative alone to determine which local minimum is the global one. In such cases, we need to evaluate the function values at different minima to compare and identify the lowest one.
```

```python
# -----------------------------------------------
# üåüNUMERICAL GRADIENT APPROXIMATION üåü
# -----------------------------------------------
# When the gradient of a function is difficult to calculate analytically,
# we approximate it using finite difference methods.
#
# The finite difference method estimates the derivative using the formula:
#
# Forward Difference Approximation:
#    f'(x) ‚âà (f(x + h) - f(x)) / h
#
# Central Difference Approximation (more accurate):
#    f'(x) ‚âà (f(x + h) - f(x - h)) / (2h)

#üîπ The gradient of a function f(x) at a point x0 is given by:

#     ‚àáf(x) ‚âà ( f(x + h) - f(x) ) / h

#   where h is a small step size.

#üîπ We will use `scipy.optimize.approx_fprime` to compute the numerical gradient.


epsilon = np.sqrt(np.finfo(float).eps)

# Point at which to calculate the gradient
x0 = np.array([1.0, 2.0])

# Calculate the gradient at the point x0
gradient = approx_fprime(x0, f, epsilon)

print("Gradient at x0:", gradient)

"""
üîπ INTERPRETATION:
   - The function f(x) = (x1 - 3)^2 + (x2 - 2)^2 is a simple quadratic function.
   - The analytical gradient is ‚àáf(x) = [ 2(x1 - 3), 2(x2 - 2) ].
   - At x0 = (1,2), we expect:
         ‚àÇf/‚àÇx1 = 2(1 - 3) = -4
         ‚àÇf/‚àÇx2 = 2(2 - 2) = 0
   - The numerical gradient should be close to [-4, 0].

üîπ KEY ADVANTAGES OF FINITE DIFFERENCES:
   ‚úÖ Works for any function where an analytical derivative is hard to compute.
   ‚úÖ Simple to implement.

üîπ KEY DISADVANTAGES:
   ‚ùå Can be computationally expensive (requires multiple function evaluations).
   ‚ùå Sensitive to step size (h too large ‚Üí inaccurate, h too small ‚Üí numerical errors).

"""


#üöÄüöÄüöÄAD vs FDM
"""
Automatic Differentiation (AD) vs Finite Difference Methods (FDM)
--------------------------------------------------------------
AD is a technique used to compute exact derivatives of a function at machine precision
by leveraging the chain rule. Unlike numerical differentiation (e.g., finite differences),
which approximates derivatives using small perturbations, AD provides precise results
without introducing truncation errors.

Finite difference methods approximate derivatives by evaluating function values
at slightly perturbed points. This introduces approximation errors, which are sensitive
to step size selection. Large step sizes lead to poor approximations, while small step sizes
can introduce floating-point errors.

Comparison:
- AD computes derivatives accurately using computational graphs.
- Finite differences approximate derivatives and suffer from errors as dimensions increase.

Below, we compare the finite difference method to the analytical derivative of
a sample multivariate function: f(x) = sum(x^2 + sin(x)). We analyze the error and computation time.
"""

# Define a sample function for testing differentiation
def multivariate_function(x):
    """ A sample function: sum of squares plus sine of each component. """
    return np.sum(x**2 + np.sin(x))

# Define the analytical derivative of the function
def analytical_derivative(x):
    """ The derivative of f(x) = x^2 + sin(x) is: df/dx = 2*x + cos(x). """
    return 2*x + np.cos(x)

# Implement finite difference method for gradient approximation
def finite_difference_derivative(f, x, h=1e-5):
    """ Computes gradient using the central finite difference method. """
    grad = np.zeros_like(x)  # Initialize gradient array
    for i in range(len(x)):
        x_plus = np.copy(x)
        x_minus = np.copy(x)
        x_plus[i] += h
        x_minus[i] -= h
        grad[i] = (f(x_plus) - f(x_minus)) / (2*h)  # Central difference formula
    return grad

# Range of dimensions to evaluate
dimensions = range(1, 101)  # Varying input size from 1D to 100D
errors = []  # Store error values
times = []   # Store computation times

# Iterate over dimensions to compare methods
for dim in dimensions:
    x = np.random.randn(dim)  # Generate random input vector

    true_grad = analytical_derivative(x)  # Compute exact derivative

    start_time = time.time()  # Start timing
    fd_grad = finite_difference_derivative(multivariate_function, x)  # Compute finite difference gradient
    elapsed_time = time.time() - start_time  # End timing
    times.append(elapsed_time)

    error = np.linalg.norm(fd_grad - true_grad)  # Compute error
    errors.append(error)

# Plot the results
plt.figure(figsize=(14, 6))

# Plot error vs. dimensions
plt.subplot(1, 2, 1)
plt.plot(dimensions, errors, marker='o', label='Finite Difference Error')
plt.xlabel('Number of Dimensions')
plt.ylabel('Error')
plt.title('Approximation Error by Dimension')
plt.grid(True)
plt.legend()

# Plot computational time vs. dimensions
plt.subplot(1, 2, 2)
plt.plot(dimensions, times, marker='o', label='Computation Time')
plt.xlabel('Number of Dimensions')
plt.ylabel('Time (seconds)')
plt.title('Computational Time by Dimension')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

"""
Interpretation:
- The error in finite difference derivatives increases with higher dimensions.
- Computation time also increases due to more function evaluations.
- AD, in contrast, remains efficient and precise in high dimensions.

This demonstrates why AD is preferred in fields like optimization.
"""
```

#**CONSTRAINED PROBLEMüîíüöß‚õî: an economic example**

```python
#Now we see an economic example of optuimization subject to a constraint
#The scipy.optimize.minimize function in Python is a versatile solver for minimization problems of both unconstrained and constrained functions.
#It provides a wide range of algorithms for different kinds of optimization problems.

# Constants
p1 = 10  # price of good 1
p2 = 20  # price of good 2
Y = 100  # income
alpha = 0.4  # elasticity for good 1
beta = 0.6   # elasticity for good 2

# Utility function: Cobb-Douglas utility function
# U(x1, x2) = x1^alpha * x2^beta
# We are minimizing the negative of the utility because scipy's 'minimize' function performs minimization
def utility(x):
    x1, x2 = x  # unpack the decision variables
    return -(x1**alpha * x2**beta)  # return the negative utility

# Budget constraint: p1 * x1 + p2 * x2 = Y
# This function returns the difference between the left-hand side (spending) and the income (Y)
def budget_constraint(x):
    x1, x2 = x  # unpack the decision variables
    return p1 * x1 + p2 * x2 - Y  # return the difference from the budget constraint

# Constraints as a dictionary: We need the budget constraint to be equal to zero
constraints = ({'type': 'eq', 'fun': budget_constraint})

# Initial guess for the optimization: We start with an equal split of the budget
initial_guess = [Y / p1, Y / p2]

# Run the optimization using scipy's 'minimize' function
result = minimize(utility, initial_guess, constraints=constraints)

# Extract optimal values of x1 and x2 (the optimal consumption bundle)
x1_opt, x2_opt = result.x
print(f"Optimal consumption bundle: x1 = {x1_opt:.2f}, x2 = {x2_opt:.2f}")

# Utility at optimal point
utility_opt = -result.fun  # since we minimized the negative utility
print(f"Optimal utility: U = {utility_opt:.2f}")

# 3D plot of the utility function: Create a mesh grid for x1 and x2 values
x1_vals = np.linspace(0, Y / p1, 100)  # range for x1
x2_vals = np.linspace(0, Y / p2, 100)  # range for x2
X1, X2 = np.meshgrid(x1_vals, x2_vals)  # mesh grid for plotting
U = X1**alpha * X2**beta  # calculate utility for each point on the grid

# Plot the utility function in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, U, cmap='viridis', edgecolor='none')

# Plot the budget constraint line in 3D: x1 and x2 points that satisfy the budget constraint
x1_budget = np.linspace(0, Y / p1, 100)  # range for x1
x2_budget = (Y - p1 * x1_budget) / p2  # calculate corresponding x2 values
ax.plot(x1_budget, x2_budget, (x1_budget**alpha * x2_budget**beta), color='r', label='Budget Line')

# Indifference curve at optimal utility: contour plot of the utility function
U_opt_surface = x1_opt**alpha * x2_opt**beta  # utility at the optimal point
indifference_curve = ax.contour(X1, X2, U, levels=[U_opt_surface], colors='k', linewidths=2)

# Labels and title for the plot
ax.set_xlabel('Good 1 (x1)')
ax.set_ylabel('Good 2 (x2)')
ax.set_zlabel('Utility (U)')
ax.set_title('Cobb-Douglas Utility Maximization with Budget Constraint')

# Show the optimal point and indifference curve on the plot
ax.scatter(x1_opt, x2_opt, utility_opt, color='r', s=100, label=f'Optimal Bundle: ({x1_opt:.2f}, {x2_opt:.2f})')

# Show the plot
plt.legend()
plt.show()

#In the final graph we can see the tangency between indifference curve and budget constraint
```

##üìå**STEP 2Ô∏è‚É£: define the AR(2) model**

```python
# üìå Understanding the AR(2) Model and Likelihood Estimation

# üéØ The AR(2) Process:
# The Autoregressive model of order 2 (AR(2)) is a time series model where the current value of the series (Y_t)
# depends on its two most recent past values (Y_{t-1} and Y_{t-2}). It is mathematically expressed as:

#   Y_t = c + œÜ1 * Y_{t-1} + œÜ2 * Y_{t-2} + Œµ_t

# where:
#   - c is a constant term.
#   - œÜ1 and œÜ2 are the autoregressive coefficients that determine how past values influence Y_t.
#   - Œµ_t is white noise with mean zero and variance œÉ¬≤ (Œµ_t ~ N(0, œÉ¬≤)).
#   - The model assumes that past values contain useful information for predicting the present.

# üìå Stationarity Condition:
# A time series is **stationary** if its statistical properties (mean, variance, and autocorrelation)
# do not change over time. For an AR(2) process to be stationary, the parameters must satisfy:

#   - œÜ2 > -1
#   - œÜ1 + œÜ2 < 1
#   - œÜ2 - œÜ1 < 1

# If these conditions hold, the process does not explode over time and remains stable.
```

##**üìåSTEP3Ô∏è‚É£.1Ô∏è‚É£: estimation of AR(2) parameters with conditional likelihood**

```python
# Likelihood Estimation:
# Likelihood estimation is a method for estimating the parameters (Œ∏ = (c, œÜ1, œÜ2, œÉ¬≤)) of a statistical model.
# The likelihood function measures how probable the observed data are, given a set of parameter values.
# In our case, we want to estimate parameters (c, œÜ1, œÜ2, œÉ¬≤) for the AR(2) model:
#
#     Y_t = c + œÜ1 Y_{t-1} + œÜ2 Y_{t-2} + Œµ_t
#
#   where Œµ_t ~ N(0, œÉ¬≤) (i.e., white noise follows a normal distribution).

# üéØ Conditional Likelihood:
# The **conditional likelihood function** assumes that the first two observations (Y1 and Y2) are given
# (we condition on them), and we calculate the probability of observing the remaining values Y3, ..., YT.

# The likelihood function is:

#   L_c(Œ∏ | Y1, Y2, ..., YT) = Œ†_{t=3}^T f(Y_t | Y_{t-1}, Y_{t-2}; Œ∏)

# üîπ This means we multiply the probabilities of each individual observation Y_t, given the past two values.
# üéØ Why Multiply Probabilities?
# - Each observation Y_t **depends on the previous two values** (Y_{t-1}, Y_{t-2}).
# - Since errors (Œµ_t) are independent, the probability of seeing the full sequence of observations
#   is the **product** of the individual conditional probabilities:

#   L_c(Œ∏) = P(Y_3 | Y_2, Y_1, Œ∏) * P(Y_4 | Y_3, Y_2, Œ∏) * ... * P(Y_T | Y_{T-1}, Y_{T-2}, Œ∏)

# üîπ This follows from the **chain rule of probability**:
#     P(A, B, C) = P(A) * P(B | A) * P(C | A, B)

# üéØ Normality Assumption:
# - We assume the **errors (Œµ_t) follow a normal distribution**: Œµ_t ~ N(0, œÉ¬≤)
# - This allows us to use the normal **probability density function (PDF)**:

#   f(Y_t | Y_{t-1}, Y_{t-2}; Œ∏) = (1 / sqrt(2œÄœÉ¬≤)) * exp(- (Y_t - c - œÜ1 Y_{t-1} - œÜ2 Y_{t-2})¬≤ / (2œÉ¬≤) )

# Substituting the normal probability density function:

#   L_c(Œ∏) = ‚àè_{t=3}^T (1 / sqrt(2œÄœÉ¬≤)) * exp(- (Y_t - c - œÜ1 Y_{t-1} - œÜ2 Y_{t-2})¬≤ / (2œÉ¬≤) )

# Taking the logarithm (log-likelihood function):

#   ‚Ñì_c(Œ∏ | Y1, Y2, ..., YT) = - (T-2)/2 * log(2œÄœÉ¬≤) - (1 / 2œÉ¬≤) * ‚àë_{t=3}^T (Y_t - c - œÜ1 Y_{t-1} - œÜ2 Y_{t-2})¬≤

# üéØ Why Use the Log-Likelihood?
# - The likelihood function involves **multiplication of probabilities**, which can result in very small numbers.
# - Taking the **logarithm** converts the product into a **sum**, making computations numerically stable.
# - The log-likelihood function is easier to differentiate, which helps in parameter estimation.

#   ‚Ñì_c(Œ∏) = - (T-2)/2 * log(2œÄœÉ¬≤) - (1 / 2œÉ¬≤) * ‚àë_{t=3}^T (Y_t - c - œÜ1 Y_{t-1} - œÜ2 Y_{t-2})¬≤

# üîπ Maximizing ‚Ñì_c(Œ∏) helps us find the best parameters (MLE estimation).


# üîπ The expression inside the exponential measures how far Y_t is from its predicted value.

# üéØ Example to Understand This Concept:
# Imagine we have the following time series data:

#   Time (t)   |  1   |  2   |  3   |  4   |  5   |
#   ---------- | ---- | ---- | ---- | ---- | ---- |
#   Y_t        | 100  | 105  | 108  | 112  | 115  |

# Assume we want to estimate the probability of observing Y3 = 108, Y4 = 112, and Y5 = 115, **given Y1 = 100 and Y2 = 105**.

# Using the AR(2) model:

#   Y_t = c + œÜ1 * Y_{t-1} + œÜ2 * Y_{t-2} + Œµ_t

# Let's assume some parameter values:
#   c = 1, œÜ1 = 0.6, œÜ2 = 0.3, and œÉ¬≤ = 4.

# We compute the predicted values:

#   Y_3_pred = 1 + (0.6 * 105) + (0.3 * 100) = 1 + 63 + 30 = 94
#   Y_4_pred = 1 + (0.6 * 108) + (0.3 * 105) = 1 + 64.8 + 31.5 = 97.3
#   Y_5_pred = 1 + (0.6 * 112) + (0.3 * 108) = 1 + 67.2 + 32.4 = 100.6

# Now, we compute the likelihood for each observation:

#   L_c(Œ∏) = ‚àè_{t=3}^5 (1 / sqrt(2œÄ * 4)) * exp(- (Y_t - Y_t_pred)¬≤ / (2 * 4))

#   L_c(Œ∏) = (1 / sqrt(8œÄ)) * exp(- (108 - 94)¬≤ / 8)  *
#            (1 / sqrt(8œÄ)) * exp(- (112 - 97.3)¬≤ / 8)  *
#            (1 / sqrt(8œÄ)) * exp(- (115 - 100.6)¬≤ / 8)

# üîπ The product of these terms gives us the **likelihood of the observed sequence** under the given parameters.

def ar_likelihood(params, data, p):
    """
    Calculate the negative (unconditional) log likelihood for an AR(p) model.

    params: list of parameters, where the first p are AR coefficients and the last is the noise variance.
    data: observed data.
    p: order of the AR model.
    """
    # Extract AR coefficients and noise variance
    c = params[0]
    phi = params[1:p+1]
    sigma2 = params[-1]

    # Calculate residuals
    T = len(data)
    residuals = data[p:] - c - np.dot(np.column_stack([data[p-j-1:T-j-1] for j in range(p)]), phi)

    # Calculate negative log likelihood
    log_likelihood = (-T/2 * np.log(2 * np.pi * sigma2) - np.sum(residuals**2) / (2 * sigma2))

    return -log_likelihood

def estimate_ar_parameters(data, p):
    """
    Estimate AR model parameters using maximum likelihood estimation.

    data: observed data.
    p: order of the AR model.
    """
    # Initial parameter guess (random AR coefficients, variance of 1)
    params_initial = np.zeros(p+2)
    params_initial[-1] = 1.0

    ## Bounds
    bounds = [(None, None)]
    # Then p AR coefficients, each bounded between -1 and 1
    bounds += [(-1, 1) for _ in range(p)]
    # The variance parameter, bounded to be positive
    bounds += [(1e-6, None)]

    # Minimize the negative log likelihood
    result = minimize(ar_likelihood, params_initial, args=(data, p), bounds=bounds)

    if result.success:
        estimated_params = result.x
        return estimated_params
    else:
        raise Exception("Optimization failed:", result.message)

#We use all our data imported from current.csv for INDPRO
#Using those 790 observations (one for each month) we calculate the parameters of AR(2) model

data = df_cleaned['INDPRO'].dropna().values
print(df_cleaned[['sasdate', 'INDPRO']]) #With this code we visualize the columns sasdate e indpro
print(data)

p = 2  # AR(2) model
params = estimate_ar_parameters(data, p)
print("Estimated parameters:", params)
```

###**COMPARISON BETWEEN CONDITIONAL AND OLS‚öñÔ∏è**

```python
def fit_ar_ols_xx(data, p):
    """
    data: observed data.
    p: order of the AR model.
    note: no constant
    """
    # Prepare the lagged data matrix
    T = len(data)
    Y = data[p:]  # Dependent variable (from p to end)
    X = np.column_stack([data[p-i-1:T-i-1] for i in range(p)])
    X = np.column_stack((np.ones(X.shape[0]), X))

    # Calculate OLS estimates using the formula: beta = (X'X)^-1 X'Y
    XTX = np.dot(X.T, X)  # X'X
    XTY = np.dot(X.T, Y)  # X'Y
    beta_hat = np.linalg.solve(XTX, XTY)  # Solve (X'X)beta = X'Y

    return beta_hat

beta_hat = fit_ar_ols_xx(data, p)
print("Estimated AR coefficients:", beta_hat)
```

## **üìåSTEP 3Ô∏è‚É£.2Ô∏è‚É£: estimation of AR(2) parameters with exact likelihood**

```python
#The Exact Likelihood method üíØüíØüíØ

# üìå Theoretical explanation o f the Exact Likelihood in AR(2) Models üìå

# ============================================================
# üîç 1. Introduction to Likelihood in Time Series Models üîç
# ============================================================

# In time series analysis, the likelihood function is essential for estimating
# the parameters of an autoregressive (AR) model. For an AR(2) process, we can define
# the likelihood function in two ways:

# 1Ô∏è‚É£ Conditional Likelihood: Assumes the first observations are given and does not
#    explicitly model their distribution.
# 2Ô∏è‚É£ Unconditional (or Exact) Likelihood: Incorporates the stationary distribution
#    of the initial values, providing a more accurate estimation.

# ============================================================
# ‚öñÔ∏è 2. Conditional vs. Unconditional Likelihood ‚öñÔ∏è
# ============================================================

# üìù --- Conditional Likelihood ---
# The conditional likelihood function assumes that the first two observations Y1 and Y2
# are fixed and does not explicitly model their distribution. It is expressed as:

#     L_c(Œ∏ | Y1, Y2, ..., YT) = Œ† f(Yt | Y_{t-1}, Y_{t-2}; Œ∏)  for t = 3 to T

# ‚úÖ Simpler computation ‚ùå Possible bias in initial values

# üìù --- Unconditional (Exact) Likelihood ---
# The exact likelihood function accounts for the distribution of the initial values Y1 and Y2
# under stationarity. It is defined as:

#     L_e(Œ∏ | Y1, Y2, ..., YT) = f(Y1, Y2; Œ∏) * Œ† f(Yt | Y_{t-1}, Y_{t-2}; Œ∏)

# ‚úÖ More accurate estimation ‚ùå More complex computation

# ============================================================
# üìä 3. Exact Likelihood for an AR(2) Model üìä
# ============================================================

# The AR(2) model is defined as:

#     Y_t = c + œÜ1 * Y_{t-1} + œÜ2 * Y_{t-2} + Œµ_t  üöÄ

# where Œµ_t is white noise with variance œÉ¬≤.

# --- üìå Mean Vector ---
# Under stationarity, the mean of Yt is given by:

#     Œº = c / (1 - œÜ1 - œÜ2)  üéØ

# The initial values (Y1, Y2) follow a bivariate normal distribution with:

# --- üßÆ Variance-Covariance Matrix ---
# The variance and first-order autocovariance of the stationary process are:

#     Œ≥(0) = œÉ¬≤ / (1 - œÜ2¬≤ - œÜ1¬≤)    (Variance) üìà
#     Œ≥(1) = œÜ1 * Œ≥(0) / (1 - œÜ2)    (First-order autocovariance) üîÑ

# The covariance matrix for the initial values is:

#     Œ£0 = | Œ≥(0)   Œ≥(1) |
#          | Œ≥(1)   Œ≥(0) |

# üì¢ where the determinant of Œ£0 is:

#     det(Œ£0) = Œ≥(0)¬≤ - Œ≥(1)¬≤

# üéØ The likelihood of the initial values (Y1, Y2) is then:

#     f(Y1, Y2; Œ∏) = (1 / (2œÄ |Œ£0|^0.5)) * exp( -0.5 * (Y - Œº)·µÄ Œ£0‚Åª¬π (Y - Œº) )  üìä

# --- üî¢ Log-Likelihood Function ---
# The total exact log-likelihood consists of two parts:
# 1Ô∏è‚É£ The likelihood of the initial values:
#
#     l_e(Y1, Y2; Œ∏) = -ln(2œÄ |Œ£0|^0.5) - 0.5 * (Y - Œº)·µÄ Œ£0‚Åª¬π (Y - Œº)
#
# 2Ô∏è‚É£ The conditional likelihood for the remaining observations:
#
#     l_e(Y3, ..., YT | Y1, Y2; Œ∏) = - (T - 2)/2 * ln(2œÄœÉ¬≤) - 1/(2œÉ¬≤) * Œ£ (Y_t - c - œÜ1Y_{t-1} - œÜ2Y_{t-2})¬≤

# üéØ Therefore, the total exact log-likelihood is:

#     L_e(Œ∏ | Y1, Y2, ..., YT) = l_e(Y1, Y2; Œ∏) + l_e(Y3, ..., YT | Y1, Y2; Œ∏)

# ============================================================
# ü§î 4. Interpretation ü§î
# ============================================================

# - ‚úÖ The exact likelihood approach accounts for the full distribution of the data,
#   providing more precise parameter estimation.
# - üîÑ The conditional likelihood is computationally simpler but may introduce bias.
# - üìâ The variance-covariance matrix ensures the process remains stationary.
# - üèÜ Maximum likelihood estimation (MLE) seeks to maximize the exact likelihood
#   to find the optimal parameters (c, œÜ1, œÜ2, œÉ¬≤).
```

```python
def ar2_exact_loglikelihood(params, y):
    """
    Calculate the exact log-likelihood for an AR(2) model.

    Parameters:
    -----------
    y : array-like
        data (T x 1)
    params : tuple or list
        Model parameters (c, phi1, phi2, sigma2)
        c: constant term
        phi1: coefficient of y_{t-1}
        phi2: coefficient of y_{t-2}
        sigma2: error variance

    Returns:
    --------
    float
        Exact log-likelihood value
    """
    # Extract parameters
    c, phi1, phi2, sigma2 = params

    # Check stationarity conditions
    if not (phi2 > -1 and phi1 + phi2 < 1 and phi2 - phi1 < 1):
        return -np.inf  # Return negative infinity if not stationary

    T = len(y)

    if T < 3:
        raise ValueError("Time series must have at least 3 observations for AR(2)")

    # Calculate the unconditional mean of the process
    mu = c / (1 - phi1 - phi2)

    # Calculate autocovariances for stationary process
    gamma0 = sigma2 / (1 - phi2**2 - phi1**2)  # Variance
    gamma1 = phi1 * gamma0 / (1 - phi2)        # First-order autocovariance

    # Create initial variance-covariance matrix
    Sigma0 = np.array([[gamma0, gamma1],
                        [gamma1, gamma0]])

    # Calculate determinant of Sigma0
    det_Sigma0 = gamma0**2 - gamma1**2

    # Calculate inverse of Sigma0
    if det_Sigma0 <= 0:  # Check for positive definiteness
        return -np.inf

    inv_Sigma0 = np.array([[gamma0, -gamma1],
                            [-gamma1, gamma0]]) / det_Sigma0

    # Initial distribution contribution (Y1, Y2)
    y_init = np.array([y[0], y[1]])
    mu_init = np.array([mu, mu])

    diff_init = y_init - mu_init
    quad_form_init = diff_init.T @ inv_Sigma0 @ diff_init

    loglik_init = -np.log(2 * np.pi * np.sqrt(det_Sigma0)) - 0.5 * quad_form_init

    # Conditional log-likelihood contribution (Y3, ..., YT | Y1, Y2)
    residuals = np.zeros(T-2)
    for t in range(2, T):
        y_pred = c + phi1 * y[t-1] + phi2 * y[t-2]
        residuals[t-2] = y[t] - y_pred

    loglik_cond = -0.5 * (T-2) * np.log(2 * np.pi * sigma2) - \
                   0.5 * np.sum(residuals**2) / sigma2

    # Total exact log-likelihood
    exact_loglik = loglik_init + loglik_cond

    ## Return the negative loglik
    return -exact_loglik

from scipy import optimize
def fit_ar2_mle(y, initial_params=None):
    """
    Fit an AR(2) model using maximum likelihood estimation

    Parameters:
    -----------
    y : array-like
        Time series data
    initial_params : tuple, optional
        Initial guess for (c, phi1, phi2, sigma2)
    """
    # Set default initial parameters if not provided
    if initial_params is None:
      # Simple initial estimates
      c_init = 0.0
      phi1_init = 0
      phi2_init = 0
      sigma2_init = np.var(y)

      initial_params = (c_init, phi1_init, phi2_init, sigma2_init)
      # Constraints to ensure positive variance

    lbnds = (-np.inf, -0.99, -0.99, 1e-6)  # Lower bounds for params
    ubnds = (np.inf, 0.99, 0.99, np.inf)     # Upper bounds for params

    bnds = optimize.Bounds(lb=lbnds, ub=ubnds)
    # Optimize
    result = optimize.minimize(
        ar2_exact_loglikelihood,
        initial_params,
        (y,),
        bounds = bnds,
        method='L-BFGS-B',
        options={'disp': False} # set to true to get more info
    )

    if not result.success:
        print(f"Warning: Optimization did not converge. {result.message}")

    # Return parameters and maximum log-likelihood
    return result.x, result.fun


Y = data
fit_ar2_mle(Y, initial_params=None)
```

```python
# First, reimport the dataset without applying the transformation code

url = 'https://www.stlouisfed.org/-/media/project/frbstl/stlouisfed/research/fred-md/monthly/current.csv?sc_lang=en&hash=80445D12401C59CF716410F3F7863B64'

df = pd.read_csv(url)

# We remove the first line with transformation codes
df_cleaned = df.drop(index=0)
df_cleaned.reset_index(drop=True, inplace=True)

# Now we convert the sasdate column into datatime format
df_cleaned['sasdate'] = pd.to_datetime(df_cleaned['sasdate'], format='%m/%d/%Y')

# With this comand we control the first rows of the dataframe
print(df_cleaned.head())

print(df_cleaned.shape)  #We control the total number of rows and columns and we get 792,127 so the dataframe has been sucsessfully imported
print(df_cleaned.columns) #We control columns names
print(df_cleaned.dtypes) #With this comand we control if the columns have been imported in the right format: we get sasdate as "datetime64" so it means that is in datatime format and the other columns are "float64" so they are in numeric format

# Extract transformation codes
transformation_codes = df.iloc[0, 1:].to_frame().reset_index()
transformation_codes.columns = ['Series', 'Transformation_Code']

#we need again to remove the first two column
df_cleaned = df_cleaned[2:]
print(df_cleaned['INDPRO'])
#Let's take the true indpro values without the log diff transformation
true_indpro = df_cleaned.loc[df_cleaned['sasdate'].dt.year == 2024, 'INDPRO'].values[:8]
print(true_indpro) #we get the first 8 values for INDPRO in 2024

def convert_logdiff_to_growth(forecasted_logdiff, last_observed_value):
    """
    Convert log-differenced forecasts to INDPRO levels.
    """
    growth_forecast = np.exp(forecasted_logdiff) - 1
    indpro_forecast = []
    last_value = last_observed_value

    for growth in growth_forecast:
        new_value = last_value * (1 + growth)
        indpro_forecast.append(new_value)
        last_value = new_value

    return np.array(indpro_forecast)

# Last two observed values up to December 2023
last_values = df_cleaned[df_cleaned['sasdate'] < '2024-01-01']['INDPRO'].values[-2:]
print(last_values)

# Log-differenced forecast obtained from the AR(2) model
forecasted_logdiff = df_forecast['Forecasted Value'].values

# Convert forecasts to INDPRO levels
indpro_forecast = convert_logdiff_to_growth(forecasted_logdiff, last_values[1])

# Create a DataFrame for the comparison
df_comparison = pd.DataFrame({
    'Month': df_forecast['Month'],
    'Forecast INDPRO': indpro_forecast,
    'Real INDPRO': true_indpro
})

print("\nComparison between forecasted and actual INDPRO:")
print(df_comparison)

# Plot the comparison
plt.figure(figsize=(10, 6))
plt.plot(df_comparison['Month'], df_comparison['Real INDPRO'], marker='o', label='Actual Value', color='blue')
plt.plot(df_comparison['Month'], df_comparison['Forecast INDPRO'], marker='x', linestyle='dashed', label='Forecast', color='red')
plt.title('Comparison between Forecasted and Actual INDPRO (2024)')
plt.xlabel('Month')
plt.ylabel('INDPRO Level')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
